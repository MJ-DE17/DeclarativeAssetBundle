```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum

spark = SparkSession.builder.getOrCreate()

df = spark.table("dev_catalog.sales_schema.silver_sales")

# Aggregate
gold_df = df.groupBy("category").agg(sum("amount").alias("total_sales"))

gold_df.write.mode("overwrite").saveAsTable("dev_catalog.sales_schema.gold_sales")
```
