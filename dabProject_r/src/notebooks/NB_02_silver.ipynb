```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.getOrCreate()

df = spark.table("dev_catalog.sales_schema.bronze_sales")

# Clean data (cast types)
df_clean = df.withColumn("amount", col("amount").cast("double"))


df_clean.write.mode("overwrite").saveAsTable("dev_catalog.sales_schema.silver_sales")
```
