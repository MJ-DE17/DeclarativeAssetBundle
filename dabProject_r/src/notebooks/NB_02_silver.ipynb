```python
dbutils.widgets.text("catalog", "")
dbutils.widgets.text("schema", "")

catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.getOrCreate()

df = spark.table(f"{catalog}.{schema}.bronze_sales")

# Clean data (cast types)
df_clean = df.withColumn("amount", col("amount").cast("double"))

table_name = f"{catalog}.{schema}.silver_sales"

df_clean.write.mode("overwrite").saveAsTable(table_name)

```

```python
spark.sql(f"SELECT * FROM {table_name}").show()
```
